```python
#importing required libraruies

import selenium
import pandas as pd
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
import time
from selenium.common.exceptions import NoSuchElementException
import requests
```

```python
driver=webdriver.Chrome("chromedriver.exe")

url = "https://www.myntra.com/reviews/4451395"
driver.get(url)
time.sleep(3)
```

```python
rat = []

ra = driver.find_elements_by_xpath("//div[@class = 'user-review-userReviewWrapper ']/div[1]/div[1]/span")
for i in ra:
    rat.append(i.text)
    
rat
```

```python
rev = []

re = driver.find_elements_by_xpath("//div[@class = 'user-review-userReviewWrapper ']/div[1]/div[2]")
for i in re:
    rev.append(i.text)
    
rev
```

```python
nlp1 = pd.DataFrame({'Reviews':rev,'rating':rat})
nlp1.to_csv('nlpdata1.csv')
```

```python
nlp1['rating'].value_counts()
```
